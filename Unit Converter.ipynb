```python
#unit converter 
#Converts various units between one another.
#The user enters the type of unit being entered, the
#type of unit they want to convert to and then the value.
#The program will then make the conversion.
from IPython.display import clear_output

def temp():
    
    while True:
        try:
            temp = input("Input the  temperature you like to convert? (e.g., 45F, 102C etc.): ")
            degree = int(temp[:-1])
            i_convention = temp[-1]
        
        except:
            print('Wrong format!')
            continue
        
        if i_convention.upper() == 'C':
            result = int(round((9 * degree) / 5 + 32))
            o_convention = 'Fahrenheiht'
        elif i_convention.upper() == 'F':
            result = int(round((degree - 32) * 5 / 9))
            o_convention = 'Celcius'
        print("\nThe temperature in", o_convention, "is", result, "degrees.")
        break

def volume_and_area():
    
    PI = 3.14
    
    while True:
        try:
            radius = float(input('Please Enter the Radius of a Cylinder: '))
            height = float(input('Please Enter the Height of a Cylinder: '))
            
            sa = 2 * PI * radius * (radius + height)
            Volume = PI * radius * radius * height
            L = 2 * PI * radius * height
            T = PI * radius * radius
        except:
            print('Wrong format!')
            continue
        
        print(f'\nThe Surface area of a Cylinder = {round(sa,2)}')
        print(f'The Volume of a Cylinder = {round(Volume,2)}')
        print(f'Lateral Surface Area of a Cylinder = {round(L,2)}');
        print(f'Top OR Bottom Surface Area of a Cylinder = {round(T,2)}')
        break

def mass():
    
    while True:
        
        try:
            flag = input("What to calculate? (m, d, v): ").lower()
            if flag == 'm':
                d = float(input("Density: "))
                v = float(input("Volume: "))
                result = d * v # mass
            elif flag == 'd':
                m = float(input("Mass: "))
                v = float(input("Volume: "))
                result = m / v # density
            elif flag == 'v':
                m = float(input("Mass: "))
                d = float(input("Density: "))
                result = m / d # volume
        except:
            print('Wrong format!')
            continue
        
        print(print(f'result is {round(result,2)}'))
        break

def replay():
    
    return input('Do you wanna use the program again? y/n: ').lower().startswith('y')
            
def main():
    
    while True:
        clear_output()
        
        prompt = input('What do you want to do? (temp,mass,volume(cylinder)): ').lower()
        
        if not prompt == 'temp' or prompt == 'volume' or prompt == 'mass':
            continue
        
        if prompt == 'temp':
            temp()
        elif prompt == 'mass':
            mass()
        elif prompt == 'volume':
            volume_and_area()
    
        if not replay():
            break

if __name__ == '__main__':
    main()
```
