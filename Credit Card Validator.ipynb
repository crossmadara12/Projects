```python
#Credit card validator
#Takes in a credit card number from a common credit card vendor(Visa, MasterCard, American Express, Discover)
#and validates it to make sure that it is a valid number

def multiply(card_number): #multiply every other digit
    number = card_number.replace(' ','')
    ret = []
    
    for x, y in enumerate(number[:-1]):
        if x % 2 == 0:
            ret.append(str(int(y)*2))
        else:
            ret.append(y)
    
    return ret

def check_digit(number): #check if there's 2 digit in the mupltiplied number and add them
    ret = []
    
    for i in number:
        if len(i) > 1:
            ret.append(str(sum(map(int, i))))
        else:
            ret.append(i)
    
    return ret

def check_validity(multiplied, number):
    ret = sum(map(int, multiplied)) + int(number[-1]) % 10
    
    return ret % 10 == 0

def replay():
    return input('You wanna check another credit card?: ').lower().startswith('y')

def main():
    
    while True:
        try:
            credit_number = input('Enter your credit card number: ')

            if not (len(credit_number) - 3 == 16):
                print('Check your credit card number! (16 digit and must have space e.g *** *** *** ***)')
                continue
            else:
                credit_multiplied = multiply(credit_number)
                credit_sum = check_digit(credit_multiplied)
                if check_validity(credit_sum, credit_number):
                    print('Valid credit card')
                else:
                    print('Invalid credit card!')
        
        except ValueError:
            print('Must be an integer!')
            continue
                
        if not replay():
            break

if __name__ == '__main__':
    main()         
```
