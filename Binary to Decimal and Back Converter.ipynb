```
from IPython.display import clear_output

#a converter to convert a decimal number to binary
#or a binary number to its decimal equivalent.

def decimal(num, base): #convert decimal to other number system
    
    if base == 2:
        print(f'Decimal to Binary is {bin(num)}')
    if base == 8:
        print(f'Decimal to Octal is {oct(num)}')
    if base == 16:
        print(f'Decimal to Hexadecimal is {hex(num)}')

def number_sys(num, base):#convert number system to decimal
    
    if base == 2:
        print(f'Binary to Decimal is {int(num, 2)}')
    if base == 8:
        print(f'Octal to Decimal is {int(num, 8)}')
    if base == 16:
        print(f'Hexadecimal to Decimal is {int(num, 16)}')

def proceed():
    
    return input('Wrong format! Do you still want to continue? y/n: ')

def replay():
    
    return input('Do you want to convert a again? y/n: ').lower().startswith('y')

def main():
    
    while True:
        clear_output()
        
        try:
            choice = int(input('How do you want it to be converted?(1.Decimal, 2.Number system): '))
            
            if not (choice == 1 or choice == 2):
                print('Choose 1 or 2!')
                continue
        except:
            print('Wrong Format, please try again!')
            continue

        while True:
            
            if choice == 1:
            
                try:
                    num = int(input('Put a decimal number: '))
                    base = int(input('Base of system you want it to be converted e.g (2 for binary): '))
                    clear_output()
                    decimal(num, base)
                    print('\nThank You!')
                    break

                except:
                    if not proceed():
                        break
                    else:
                        continue

            else:
                if choice == 2:
                
                    try:
                        num = input('Enter the number: ')
                        base = int(input('Base of the number: '))
                        clear_output()
                        number_sys(num, base)
                        print('\nThank You!')
                        break

                    except:
                        if not proceed():
                            break
                        else:
                            continue
                      
        if not replay():
            break

if __name__ == '__main__':
    main()
```
